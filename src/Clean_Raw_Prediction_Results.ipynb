```python
df = pd.DataFrame.from_dict(temp, orient="index")
df.head()
```

```python
percentages = df['values'].apply(pd.Series).apply(lambda x: round(x * 100, 2))
col_names = [
    'abstract',
     'bands',
     'circle',
     'figure',
     'gestural',
     'grid',
     'landscape',
     'portrait',
     'rectangle',
     'repetition',
     'shape',
     'square',
     'still_life',
     'symmetry'
]
percentages.columns = col_names
percentages['label'] = df['label']
percentages.head()
```

# use apply(pd.Series) for each column in this new dataframe

```python
cols_unpack = df.columns.tolist()[:-2]
cols_unpack
```

```python
color_info = pd.DataFrame()

for col in cols_unpack:
    temp = df[col].apply(pd.Series)
    color_info = pd.concat([color_info, temp], axis=1)
    
color_info.shape
```

```python
color_info.head()
```

```python
full_df = pd.concat([percentages, color_info], axis=1)
full_df.head()
```

```python
cols_to_drop = [
    'min_Blue',
    'max_Blue',
    'median_Blue',
    
    'min_Green',
    'max_Green',
    'median_Green',
    
    'min_Red',
    'max_Red',
    'median_Red',
    
    'min_Hue',
    'max_Hue',
    'median_Hue',
    
    'min_Saturation',
    'max_Saturation',
    'median_Saturation',
    
    'min_Value',
    'max_Value',
    'median_Value',
]

final_df = full_df.drop(columns=cols_to_drop)
final_df.shape
```
